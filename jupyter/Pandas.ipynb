# Data wrangling

```python
# 1. añadir newspaper_uid al DataFrame

semana['newspaper_uid'] = 'semana'

semana
```

```python
# 2. obtener el host

from urllib.parse import urlparse

semana['host'] = semana['url'].apply(lambda url: urlparse(url).netloc)

semana
```
